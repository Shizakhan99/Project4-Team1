```python
import matplotlib.pyplot as plt

# Models and their respective metrics
models = [
    "Logistic Regression",
    "K-Nearest Neighbor",
    "Naive Bayes",
    "Decision Tree",
    "Random Forest",
    "Deep Neural Network"
]

accuracy = [76, 76, 72, 96, 92, 71]
recall = [68, 71, 76, 0, 13, 74]

# Plot for Accuracy
plt.figure(figsize=(10, 6))
plt.bar(models, accuracy, color='lightblue', edgecolor='black')
plt.title('Accuracy Comparison of Models')
plt.ylabel('Accuracy (%)')
plt.xlabel('Models')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot for Recall
plt.figure(figsize=(10, 6))
plt.bar(models, recall, color='salmon', edgecolor='black')
plt.title('Recall Comparison of Models')
plt.ylabel('Recall (%)')
plt.xlabel('Models')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
