```python
import pandas as pd
from pathlib import Path
```

```python
# Read salary data
file_path = Path("Resources/Stroke_dataset.csv")
df_stroke = pd.read_csv(file_path)

# Display sample data
df_stroke.head()
```

```python
df_stroke_cleaned = df_stroke.dropna(subset=['bmi'])
df_stroke_cleaned.info()
```

## Preprocess categorical data

## Logistic Regression

```python
X = df_stroke_cleaned.drop(columns=['stroke'], inplace=False)
y = df_stroke_cleaned['stroke']
```

```python
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

## Random Forest


```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
# Step 1: Load the dataset
file_path = 'Resources/Stroke_dataset.csv'
df_stroke_cleaned = pd.read_csv(file_path)
```

```python
# Step 2: Handle missing values for 'bmi' using mean imputation
imputer = SimpleImputer(strategy='mean')
df_stroke_cleaned['bmi'] = imputer.fit_transform(df_stroke_cleaned[['bmi']])
```

```python
# Step 3: Encode categorical variables
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
encoders = {col: LabelEncoder() for col in categorical_cols}
for col, encoder in encoders.items():
    df_stroke_cleaned[col] = encoder.fit_transform(df_stroke_cleaned[col])
```

```python
# Step 4: Define features and target variable
X = df_stroke_cleaned.drop(columns=['id', 'stroke'])  # Exclude 'id' column
y = df_stroke_cleaned['stroke']
```

```python
# Step 5: Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

```

```python
# Step 6: Train a Random Forest Classifier
rf_clf = RandomForestClassifier(
    n_estimators=100,    # Number of trees
    max_depth=10,        # Maximum depth of each tree
    random_state=42,
    class_weight="balanced"  # Address class imbalance
)
rf_clf.fit(X_train, y_train)
```

```python
# Step 7: Evaluate the model
y_pred = rf_clf.predict(X_test)
report = classification_report(y_test, y_pred)

# Display the classification report
print(report)
```
